## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.
#check missing values
df.isnull().sum()
```

```python
#check duplicates
df.duplicated().sum()
```

```python
#remove cancelled orders
#Cancelled invoices start with "C"
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
#Remove negative quantity
df = df[df['Quantity'] > 0]
```

```python
#remove zero price
df = df[df['UnitPrice'] > 0]
```

```python
#Create total price column
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

**Chart - 1 Monthly Sales Trend(Time vs Revenue)**

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

df['Month'] = df['InvoiceDate'].dt.to_period('M')

monthly_sales = df.groupby('Month')['TotalPrice'].sum()

monthly_sales.plot(figsize=(10,5))
plt.show()
```

##### 1. Why did you pick the specific chart?

•	Shows seasonality.

•	Identify peak sales month.

•	Detect growth trend.

##### 2. What is/are the insight(s) found from the chart?

Sales increased towards the end of the year, indicating seasonal buying behavior.

```python
monthly_sales = df.groupby(df['InvoiceDate'].dt.to_period('M'))['TotalPrice'].sum()

monthly_sales.plot(figsize=(10,5))

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

A heatmap displays correlation values using colors.

Dark color → strong relationship

Light color → weak relationship

annot=True means show the correlation numbers on the chart.
